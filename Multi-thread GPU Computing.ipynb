Inspired by exercice 2 during course,  
the goal of this exercice is to shift all values of a 4x4 array to the right. With "a" as the input array and "b" as the output array, first value of "a" corresponds to second value of "b", second value of "a" corresponds to third value of "b" etc...   
Exception: first value of "b" corresponds to last value of "a"  

Expected result:
 
######## a #########  
[[59 85  1 40]  
[11 57 36 34]  
[22 91 98  5]  
[79 61 71 26]]    
######## b #########  
[[26 59 85  1]  
[40 11 57 36]  
[34 22 91 98]  
[ 5 79 61 71]]  

**Version Python sur CPU:**

```python
# create 4x4 array of random integers from 1 to 99
a = numpy.random.randint(low=1, high=100, size=(4,4))
a = a.astype(numpy.int32)
b = numpy.empty_like(a)

for i in range(a.shape[0]):
  for j in range(a.shape[1]):
    if j == 0:
      b[i,j] = a[i-1,j-1]
    else:
      b[i,j] = a[i,j-1]

print(a)
print(' ')
print(b)
```

```python
import numpy
import sys

import pycuda.autoinit
import pycuda.driver as drv
from pycuda.compiler import SourceModule
```

> CUDA KERNEL

```python
mod = SourceModule("""
__global__ void shiftme(int *a[], int *b[])
  {
    int idx = blockIdx.x * blockDim.x + threadIdx.x;
    const int size = sizeof(a);

    b[idx] = a[idx-1];
  }
""")
```

> PYTHON CODE

```python
print("############ ",drv.Device(0).name(), "############")

# create 4x4 array of random integers from 1 to 99
a = numpy.random.randint(low=1, high=100, size=(4,4))
a = a.astype(numpy.int32)

# create a new empty array with same shape as array a
b = numpy.empty_like(a)

size = a.size
print(a.shape, a.dtype, a.nbytes, int(a.size))

# allocate memory on device with same size of array a
a_gpu = drv.mem_alloc(a.nbytes)
b_gpu = drv.mem_alloc(b.nbytes)
size_gpu = drv.mem_alloc(sys.getsizeof(size))

# copy a content to device memory ("htod" means "host to device")
drv.memcpy_htod(a_gpu, a)
drv.memcpy_htod(b_gpu, b)
drv.memcpy_htod(size_gpu, size.to_bytes(2, byteorder='big'))

# invoke cuda kernel called doublify
shiftme = mod.get_function("shiftme")
shiftme(a_gpu, b_gpu, block=(16,1,1))


# copy back results from device ("dtoh" means device to host)
drv.memcpy_dtoh(b, b_gpu)

print("######## a #########")
print (a)
print("######## b #########")
print (b)
```

```python
img_path = "images/img.jpg"
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
img = img.astype(numpy.int32)
maximized = numpy.empty_like(img)
minimized = numpy.empty_like(img)
height = img.shape[0]
width = img.shape[1]

brushsize = 3

for i in range(height):
  for j in range(width):

    # Selecting the neighberhood
    left = j-brushsize
    if left < 0 : left = 0
    elif left > width :  left = width
    right = j+brushsize+1
    if right < 0 : right = 0
    elif right > width :  right = width
    bottom = i-brushsize
    if bottom < 0 : bottom = 0
    elif bottom > height :  bottom = height
    top = i+brushsize+1
    if top < 0 : top = 0
    elif top > height :  top = height

    neighberhood = img[bottom:top,left:right]

    # print(neighberhood)

    # Finding max and min
    maximized[i,j] = neighberhood.max()
    minimized[i,j] = neighberhood.min()

# result
#Show Images
plt.rcParams['figure.figsize'] = [20, 10]
fig, (ax0,ax1, ax2) = plt.subplots(1, 3)

# Show Original
ax0.set_title("Original")
ax0.imshow(img, cmap=plt.get_cmap('gray'), vmin = 0, vmax = 255)

# Show max
ax1.set_title("Maximized")
ax1.imshow(maximized, cmap=plt.get_cmap('gray'), vmin = 0, vmax = 255)

# Show min
ax2.set_title("Minimized")
ax2.imshow(minimized, cmap=plt.get_cmap('gray'), vmin = 0, vmax = 255);
```
